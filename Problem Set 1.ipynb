### Q.6

````python
nums = [1,2,2,3,4,4,5]
seen = []
def contains(seen,num):
    for it in seen:
        if(it == num):
            return True
    return False
for num in nums:
    if(not contains(seen,num)):
        seen.append(num)
print(seen)
````

### Q.9

````python
def is_prime(n):
    if(n < 2):
        return False
    for i in range(2,n):
        if(n%i == 0):
            return False
    return True
````

### Q.10

````python
def factorial(n):
    if(n < 0):
        return "Domain error"
    factorial = 1
    while(n > 0):
        factorial*=n
        n-=1
    return factorial
````
