Batch Gradient Descent vs Stochastic Gradient Descent

Implement SGD: update weights and bias after each row instead of after the full dataset.

Compare final weights and bias with batch gradient descent on the same dataset.

`````python
X = [[1,2],[3,4],[5,6]]
Y = [5,11,17]
m,n = len(X),len(X[0])

x_train = X[:round(0.6*m)]
x_test = X[round(0.6*m):]
y_train = Y[:round(0.6*m)]
y_test = Y[round(0.6*m):]

thetas = [0.1 for _ in range(-1,n)] #adding the bias before as x_0
x_train = [[1]+row for row in x_train]
x_test = [[1]+row for row in x_test]
def transpose(vector):
    return [[x] for x in vector]
def hypothesis(X, theta):
    return [sum(X[i][j] * theta[j] for j in range(len(theta))) for i in range(len(X))]
def gradient_descend(X, Y, theta, alpha,epochs):
    for _ in range(epochs):
        #h_theta(X) - Y is the error
        y_hat = hypothesis(X,theta)
        errors = [y_hat[i] - Y[i] for i in range(len(Y))]
        #now this error time X is the J(theta)
        change = [sum([errors[i]*X[i][j] for i in range(len(X))]) for j in range(len(X[0]))] 
        theta = [theta[i] - (alpha/len(Y)) * change[i] for i in range(len(theta))]
    return theta
def stochastic_gradient_descend(X, Y, theta, alpha, epochs):
    for _ in range(epochs):
        for i in range(len(Y)):
            y_hat = sum(X[i][j] * theta[j] for j in range(len(theta)))
            error = y_hat - Y[i]
            change = [X[i][j] * error for j in range(len(theta))]
            theta = [theta[j] - alpha * change[j] for j in range(len(theta))]
    return theta

print(gradient_descend(x_train,y_train,thetas,0.01,10000))
stochastic_gradient_descend(x_train,y_train,thetas,0.1,10000)
`````

# 📌 Mini-batch, Momentum & Adaptive Learning Notes

<details>
<summary>1️⃣ Gradient Noise in Mini-batch</summary>

- Mini-batch gradient:  
  \[
  g_B = \frac{1}{|B|} \sum_{i \in B} \nabla_\theta L(x_i, y_i)
  \]
- Each mini-batch gives a slightly different vector → "noisy gradient"
- Noise = deviation from the true full-batch gradient
- Smaller batch → higher variance → more zig-zag
- Larger batch → lower variance → smoother update
</details>

<details>
<summary>2️⃣ Momentum Intuition</summary>

- Momentum: running average of previous gradients  
  \[
  v_t = \beta v_{t-1} + (1-\beta) g_B
  \]
  \[
  \theta_t = \theta_{t-1} - \alpha v_t
  \]
- Think: **net vector of previous mini-batch directions**
- Smooths out noise → accelerates convergence along consistent directions
- Higher β → more trust in previous direction → faster travel along valley
</details>

<details>
<summary>3️⃣ Adaptive Learning Rates</summary>

- Optimizers like Adam/RMSProp track gradient variance per dimension  
- Scales step sizes inversely with variance:  
  - High noise → smaller step  
  - Stable direction → bigger step
- Combines with momentum:
  - Momentum → direction  
  - Adaptive rate → magnitude
</details>

<details>
<summary>4️⃣ Practical Takeaways</summary>

- Mini-batch GD = standard in almost all deep learning
- SGD = mostly for online / streaming data or tiny datasets
- Batch GD = rare, only for small datasets
- Momentum + adaptive optimizer → “AI for AI”:
  - Uses history + variance to adjust updates automatically
- Visual intuition: each mini-batch is a vector, momentum = average vector, adaptive scaling = step adjustment along each axis
</details>

Test Cases:

Perfect prediction → 1.0

Constant prediction → 0.0



`````python
y_true = [1,2,3]
y_pred = [1.1,1.9,3.2]
def mae(y_true,y_pred):
    error = [abs(y_pred[i]-y_true[i]) for i in range(len(y_true))]
    return sum(error)/len(error)
def r2_score(y_true,y_pred):
    error = [(y_pred[i]-y_true[i])**2 for i in range(len(y_true))]
    mean = sum(y_true)/len(y_true)
    sd = [(x-mean)**2 for x in y_true]
    return 1-(sum(error)/sum(sd))
mae(y_true,y_pred)
r2_score(y_true,y_pred)
`````
