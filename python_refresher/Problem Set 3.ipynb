### Q1. Vector Addition
```python

```

````python
a = [1,2,3]
b = [4,5,6]
c = [a[i]+b[i] for i in range(len(a))]
print(c)
````

### Q5. Matrix Transpose

Input:
```python
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]
```
👉 Output:
```python
[
 [1, 4],
 [2, 5],
 [3, 6]
]
```

````python
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]
transpose = [[0]*len(matrix) for _ in range(len(matrix[0]))]
for i in range(len(matrix)):
    for j in range(len(matrix[0])):
        transpose[j][i] = matrix[i][j]
print(transpose)
````

### Q8. Standardization (VERY ML IMPORTANT)

Formula:

(x - mean) / std

👉 Compute for a list

````python
a = [1,2,3,4]
mean = sum(a)/len(a)
std = (sum((x-mean)**2 for x in a)/len(a))**0.5
a = [((x-mean)/std) for x in a]
print(a)
````

### Q11. Filter Rows

👉 Keep rows where second value > 250

````python
data = [
    [1, 100],
    [2, 200],
    [3, 300]
]

data = [x[1] for x in data if x[1] > 250]
print(data)
````

### Q12. Add Bias Term (SUPER IMPORTANT for ML)
