## Phase 1 — Advanced Vector Operations

### Q.1 Vector Norm

Compute L2 norm (Euclidean length) of a vector:

a = [3, 4]

output: 5

Formula: $ ||a|| = \sqrt{\sum{𝑎_{𝑖}}^{2}} $ 


```python
a = [3,4]
L2 = sum(x*x for x in a)**0.5
print(L2)
```

### Vector Mean & Centering

Compute the mean of a vector and center it: x_centered = x - mean(x)

```python
x = [3,4]
mean = sum(x)/len(x)
x_centered = [(x1-mean) for x1 in x]
print(x_centered)
```

### Vector Projection

Project vector a onto vector b:
Formula: $ proj_{b}(a) = \frac{𝑎 \cdot 𝑏}{||b||^{2}} \cdot b $

```python
a = [2, 3, 4]
b = [4, 5, 6]
dot_product = sum([a[i]*b[i] for i in range(len(a))])
L2 = sum(x*x for x in b)**0.5
proj_b_a = [dot_product/(L2**2)*b1 for b1 in b]
print(proj_b_a)
```

## Phase 2 — Matrix Operations

### Matrix Addition & Subtraction

Implement functions: matrix_add(A,B) and matrix_sub(A,B)


```python
def matrix_add(A,B):
    return [[A[i][j]+B[i][j] for j in range(len(A[0]))] for i in range(len(A))]
def matrix_sub(A,B):
    return [[A[i][j]-B[i][j] for j in range(len(A[0]))] for i in range(len(A))]
matrix_sub([[1,2],[1,3]],[[1,3],[1,2]])
```

### Matrix-Matrix Multiplication

Standard definition using loops

```python
def matrix_mult(A, B):
    return [[A[i][j]*B[j][i] for j in range(len(A[0]))] for i in range(len(A))] 
matrix_mult([[1,0],[0,1]],[[1,0],[0,1]])
```

### Determinant of 2x2 Matrix

Compute: det([[a,b],[c,d]]) = a*d - b*c

```python
def determinant(A):
    return (A[0][0]*A[1][1] - A[1][0]*A[0][1])
determinant([[1,0],[0,1]])
```

### Identity Matrix

Generate n x n identity matrix without libraries

```python
def identity(n):
    return [[1 if i==j else 0 for j in range(n)  ] for i in range(n)]
identity(3)
```

### Determinant of any matrix

```python
def determinant(A):
    return A[0][0]*A[1][1] - A[0][1]*A[1][0]

def determinant_any(A):
    n = len(A)
    m = len(A[0])
    if n != m:
        raise ValueError("Matrix must be square")
    if n == 2:
        return determinant(A)
    det = 0
    for j in range(n):
        minor = [row[:j] + row[j+1:] for row in A[1:]]
        det += ((-1) ** j) * A[0][j] * determinant_any(minor)
    return det
print(determinant_any([[1,0,0],[0,1,0],[0,0,1]]))  # Output: 1
```
