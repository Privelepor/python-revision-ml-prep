### Q1. Frequency Counter

Given:

```python
s = "machinelearning" 
```

👉 Create a dictionary that stores frequency of each character.

````python
s = "machinelearning"
freq = {}
for c in s:
    if(c in freq):
        freq[c]+=1
    else:
        freq[c] = 1

print(freq)
````

### Q2. Student Marks Analysis

```python
marks = {
    "Alice": 85,
    "Bob": 72,
    "Charlie": 90
}
```


Tasks:

Print the student with highest marks

Compute average marks

Add a new student "David": 78"

````python
marks = {
    "Alice": 85,
    "Bob": 72,
    "Charlie": 90
}

highest = -1
student_with_highest = ''
number_of_student = 0
total = 0
for student,mark in marks.items():
    if(mark > highest):
        highest = mark
        student_with_highest = student
    number_of_student+=1
    total+=mark
    
print(student_with_highest)
print(f"{total / number_of_student:.2f}")
marks["David"] = 78
````

### Q3. Common Elements
```python
    a = [1, 2, 3, 4]
    b = [3, 4, 5, 6]
    
```
👉 Find common elements using:

loops

sets

````python
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
seen = []
def contains(seen,n):
    for num in seen:
        if(num == n):
            return True
    return False
for n in a:
    if not contains(seen,n):
        seen.append(n)

for n in b:
    if contains(seen,n):
        print(f"{n} is common")

set_a = set(a)

for n in b:
    if n in set_a:
        print(f"{n} is common")
````

### Q9. Normalize Data
```python
nums = [10, 20, 30, 40, 50]
```
👉 Normalize values between 0 and 1:
Formula:
```python
(x - min) / (max - min)
```

````python
nums = [10,20,30,40,50]
minimum = min(nums)
maximum = max(nums)

nums = [(x - minimum)/(maximum - minimum) for x in nums]
print(nums)
````

### Q10. Simple Data Filter
```python
data = [
    {"name": "A", "age": 17},
    {"name": "B", "age": 22},
    {"name": "C", "age": 19}
]
```

👉 Return only people with age ≥ 18

````python
data = [
    {"name": "A", "age": 17},
    {"name": "B", "age": 22},
    {"name": "C", "age": 19}
]
data = [d for d in data if d["age"] >= 18]
print(data)
````
