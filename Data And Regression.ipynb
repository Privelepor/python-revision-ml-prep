## Part B — Probability & Statistics Foundations

Implement mean, variance, covariance, correlation coefficient for two vectors.\
Simulate sampling from a normal distribution and plot histogram (if using matplotlib).

```python
import numpy as np
X = np.array([[12,43,123,2424,1,123],[67,4,3,98,231,1],[178,42,32,76,43,21]])
Y = np.array([1,2,1])
prob_X = X / X.sum(axis=1, keepdims=True)
unique, counts = np.unique(Y, return_counts=True)
prob_Y = counts / counts.sum()
print(prob_X)
print(prob_Y)
```

```python
X = np.array([[12,43,123,2424,1,123],[67,4,3,98,231,1],[178,42,32,76,43,21]])
Y = np.array([1,2,1])

mean_X = np.mean(X,axis = 0)
mean_Y = np.mean(Y,axis = 0)
print(mean_X)
print(mean_Y)
print("Variances")
variance_X = np.var(X,axis = 0)
variance_Y = np.var(Y,axis = 0)
print(variance_X)
print(variance_Y)


print("Covariances")
covariance_X = np.cov(X,rowvar=False)
covariance_Y = np.cov(Y,rowvar=False)
print(covariance_X)
print(covariance_Y)
```

## Part C — Mini ML Pipeline with NumPy
Re-implement linear regression using full vectorization.
Implement batch gradient descent and mini-batch gradient descent.
Compute MAE, MSE, and R² metrics using NumPy.

```python
import numpy as np
import pandas as pd


# Set seed for reproducibility
np.random.seed(42)

# Generate 100 samples with 2 features
X = np.random.randn(100, 2) * 10  # random numbers, scale up
Y = 3*X[:,0] + 2*X[:,1] + 5 + np.random.randn(100)*5  # linear relation + noise

# Introduce some missing values
X[5, 0] = np.nan
X[20, 1] = np.nan

# Convert to DataFrame for easier handling
X_df = pd.DataFrame(X, columns=["feature1", "feature2"])
Y_df= pd.Series(Y, name="target")

# print(X_df.head())
# print(Y_df.head())
mean = X_df.mean()
X_df = X_df.fillna(mean)

nonEmpty = Y_df.notna();
Y_df = Y_df[nonEmpty]
X_df = X_df[nonEmpty]

X = X_df.to_numpy()
Y = Y_df.to_numpy()

indices = np.random.permutation(len(X))

X = X[indices]
Y = Y[indices]
X_train = X[:round(0.8*len(X))]
Y_train = Y[:round(0.8*len(X))]
X_test = X[round(0.8*len(X)):]
Y_test = Y[round(0.8*len(X)):]
        
m,n = X_train.shape
thetas = np.array([0.01]*(n+1))
X_train = np.insert(X_train,0,1.0,axis = 1)
X_test = np.insert(X_test,0,1.0,axis = 1)
class LinearRegression:
    
    def __init__(self, thetas, X_train, X_test, Y_train, Y_test, alpha=0.001, epochs=1000):
        self.thetas = thetas
        self.X_test = X_test
        self.X_train = X_train
        self.Y_test = Y_test
        self.Y_train = Y_train
        self.alpha = alpha
        self.epochs = epochs
        
    def predict(self, X):
        return X @ self.thetas
        
    def fit(self):
        for epoch in range(self.epochs):
            Y_hat = self.predict(self.X_train)
            
            if epoch % 100 == 0:
                print(f"After {epoch} iterations:")
                print("Thetas: ",self.thetas)
                print("Mean Squared Error: ",self.mse(self.Y_train, Y_hat))
                
            self.thetas -= self.alpha * self.gradient()

    def gradient(self):
        m = len(self.X_train)
        return (1/m) * np.dot(self.X_train.T, (self.predict(self.X_train) - self.Y_train))
    
    def mse(self, Y, Y_expected):
        return np.mean((Y - Y_expected) ** 2)
model = LinearRegression(thetas,X_train,X_test,Y_train,Y_test,0.01,1000)
model.fit()
print(model.predict(X_test))
print(Y_test)
print(thetas)
```
